# Q1.	When working with data in a business context it is important to be able to identify the type of data you are working with in relation to the business problem. Please categorize each data field by its segmentation parameter type: “Demographic,” “Purchasing,” “Behavioral,” “Attitude,” or “Other.”

```python
data_fields = [
    "age", "gender", "income", "address", "product_purchased", "purchase_frequency",
    "website_visits", "time_spent_on_site", "customer_satisfaction", "brand_loyalty"
]

segmentation_mapping = {
    "Demographic": ["age", "gender", "income", "address"],
    "Purchasing": ["product_purchased", "purchase_frequency"],
    "Behavioral": ["website_visits", "time_spent_on_site"],
    "Attitude": ["customer_satisfaction", "brand_loyalty"],
}

def categorize_data_fields(data_fields, segmentation_mapping):
    categorized_data_fields = {}
    
    for field in data_fields:
        for segment, fields in segmentation_mapping.items():
            if field in fields:
                categorized_data_fields[field] = segment
                break
        else:
            categorized_data_fields[field] = "Other"
            
    return categorized_data_fields

categorized_data_fields = categorize_data_fields(data_fields, segmentation_mapping)

for field, segment in categorized_data_fields.items():
    print(f"{field}: {segment}")
```

# Q3.	Create a table for each attribute (e.g., Gender) and record the percentage of responses for each answer (e.g., % of respondents answering "male" and % answering “female”).

```python
import pandas as pd

data = pd.read_excel('C:/Users/Prate/OneDrive/Downloads/Tivo Survey Case Data(1)(1).xlsx')


attributes = ['Gender', 'Marital Status', 'Work Status', 'Education']

tables = {}

for attribute in attributes:
    counts = data[attribute].value_counts()
    percentages = (counts / len(data)) * 100
    tables[attribute] = percentages

for attribute, table in tables.items():
    print(f"\nTable for {attribute}:")
    print(table)
```

# Q4.	Create a correlation matrix for the data set and describe the relationship you observe between the following variables:

a.	Age vs. Purchasing Frequency
b.	Age vs. TV Viewing
c.	Annual Income vs. TV Viewing
d.	Education vs. Favourite Feature
e.	Monthly Electronic Spending vs Monthly Household Spending
f.	Are there any other relationships from the matrix are any high enough to make one of the attributes redundant? In other words, you can explain any of the attributes by simply saying "it behaves like attribute X?”

```python
import pandas as pd
import numpy as np


data = pd.read_excel('C:/Users/Prate/OneDrive/Downloads/Tivo Survey Case Data(1)(1).xlsx')


education_levels = {'none': 0, 'HS': 1, 'BA': 2, 'MA': 3, 'PhD': 4}
data['Education'] = data['Education'].map(education_levels)


features = data['Favorite feature'].unique()
feature_mapping = {feature: i for i, feature in enumerate(features)}
data['Favorite feature'] = data['Favorite feature'].map(feature_mapping)


corr_matrix = data.corr()


print("Correlation Matrix:")
print(corr_matrix)


relationships = [
    ('Age', 'Purchasing Frequency (every x months)'),
    ('Age', 'TV Viewing (hours/day)'),
    ('Annual Income (x1000 $)', 'TV Viewing (hours/day)'),
    ('Education', 'Favorite feature'),
    ('Monthly Electronics Spend', 'Monthly Household Spend')
]

for var1, var2 in relationships:
    correlation = corr_matrix.loc[var1, var2]
    print(f"\nCorrelation between {var1} and {var2}: {correlation:.2f}")


redundancy_threshold = 0.9
redundant_pairs = np.where((corr_matrix > redundancy_threshold) & (corr_matrix < 1))
redundant_attributes = [(corr_matrix.index[x], corr_matrix.columns[y]) for x, y in zip(*redundant_pairs)]
if len(redundant_attributes) > 0:
    print("\nRedundant attribute pairs:")
    for attribute_pair in redundant_attributes:
        print(attribute_pair)
else:
    print("\nNo redundant attributes found.")
```

# Q5.	Answer the following questions:

a.	How many married men who are early adopters have monthly electronics spend high enough that they can afford to purchase a TiVo for $499 and still be able to spend more on electronics in the next two years? 
b.	How many women with an education level of MA or PhD are making purchasing decisions for electronics without discussing them with a spouse, either because they are single or because they are making purchasing decisions without the involvement of their spouses? 
c.	How many early adopters purchase electronics at least once every year, and do so in stores that specialize in electronics? 
d.	How many seniors (over the age of 65) spend more than six hours a day watching TV? What is their income range? What is their average annual income? 


```python
import pandas as pd


data = pd.read_excel('C:/Users/Prate/OneDrive/Downloads/Tivo Survey Case Data(1)(1).xlsx')


condition_5a = (
    (data['Gender'] == 'male') &
    (data['Marital Status'] == 'married') &
    (data['Technology Adoption'] == 'early') &
    (data['Monthly Electronics Spend'] * 24 > 499)
)
result_5a = data[condition_5a].shape[0]
print(f"5a. Number of married men who are early adopters and can afford TiVo: {result_5a}")


condition_5b = (
    (data['Gender'] == 'female') &
    ((data['Education'] == 'MA') | (data['Education'] == 'PhD')) &
    (data['Purchasing Decision-maker'] == 'single')
)
result_5b = data[condition_5b].shape[0]
print(f"5b. Number of women with MA or PhD making solo purchasing decisions: {result_5b}")


condition_5c = (
    (data['Technology Adoption'] == 'early') &
    (data['Purchasing Frequency (every x months)'] <= 12) &
    (data['Purchasing Location'] == 'specialty stores')
)
result_5c = data[condition_5c].shape[0]
print(f"5c. Number of early adopters who purchase electronics annually in specialty stores: {result_5c}")


condition_5d = (
    (data['Age'] > 65) &
    (data['TV Viewing (hours/day)'] > 6)
)
result_5d = data[condition_5d].shape[0]
income_range_5d = (
    data.loc[condition_5d, 'Annual Income (x1000 $)'].min(),
    data.loc[condition_5d, 'Annual Income (x1000 $)'].max()
)
average_income_5d = data.loc[condition_5d, 'Annual Income (x1000 $)'].mean()
print(f"5d. Number of seniors spending over 6 hours a day watching TV: {result_5d}")
print(f"     Their income range: {income_range_5d}")
print(f"     Their average annual income: {average_income_5d:.2f} (x1000 $)")
```
